```python
import numpy as np
import pandas as pd
from pandas import DataFrame, Series
import tensorflow as tf
from tensorflow import keras
import pickle
```

```python
npz_train = np.load('Absenteeism_data_train_v2.npz')
x_train = npz_train['inputs'].astype(np.float)
y_train = npz_train['target'].astype(np.int)

npz_test = np.load('Absenteeism_data_test_v2.npz')
x_test, y_test = npz_test['inputs'].astype(np.float), npz_test['target'].astype(np.int)
```

# Model

data, model, loss function and optimization

```python
y_train.shape
```

```python
imput_layer = 544
hidden_layer = 200
output_layer = 2

model = tf.keras.Sequential([tf.keras.layers.Dense(hidden_layer,activation='relu'),
                             tf.keras.layers.Dense(hidden_layer, activation='relu'),
                             tf.keras.layers.Dense(hidden_layer, activation='relu'),
                             tf.keras.layers.Dense(output_layer,activation='softmax')])

opt = tf.keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=opt,loss='sparse_categorical_crossentropy',metrics=['accuracy'])

Batch = 60
max_epoch = 100
early_stopping = tf.keras.callbacks.EarlyStopping(patience=2)

model.fit(x_train, y_train, batch_size=Batch, epochs=max_epoch, callbacks=[early_stopping],
         verbose=2,validation_split=0.1)
```

# Test

```python
loss,accuracy = model.evaluate(x_test,y_test,verbose=0)
```

```python
print('{0:.2f}'.format(loss))
print('{0:.2f}%'.format(accuracy*100))
```

```python
predictions = model.predict_classes(x_test)
predictions.round(3)
```
